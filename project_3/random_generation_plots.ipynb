```python
import random
import math
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
n = 20
R = 5.0
d = 10

M = random.randint(-100, 100), random.randint(-100, 100)
```

```python
circumference_points = []
for _ in range(n):
    a = random.random() * 2 * math.pi
    r = R

    x = r * math.cos(a) + M[0]
    y = r * math.sin(a) + M[1]
    circumference_points.append((x, y))
```

```python
circumference = Polygon(circumference_points)
```

```python
import matplotlib.pyplot as plt
figure, axes = plt.subplots()
draw_circle = plt.Circle(M, R, fill=False)
plt.scatter(*zip(*circumference_points))
plt.show()
```

```python
inner_points = []
while len(inner_points) < (int(d * math.pi * R**2)):
    a = random.random() * 2 * math.pi
    r = R * math.sqrt(random.random())
    
    x = r * math.cos(a) + M[0]
    y = r * math.sin(a) + M[1]
    if circumference.contains(Point(x,y)):
        inner_points.append((x, y))
```

```python
import matplotlib.pyplot as plt
plt.scatter(*zip(*(circumference_points + inner_points)))
plt.show()
```
